```python
DATADIR = "/content/drive/MyDrive/breastcancerdataset/training/benign/8863_idx5_x1001_y1001_class0.png"
```

```python
model="/content/drive/MyDrive/breastcancerdataset/breastcancerprediction1.h5"
loaded_model=tf.keras.models.load_model(model)
```

```python
img = tf.keras.utils.load_img(
    DATADIR, target_size=(180,180)
)
img_array = tf.keras.utils.img_to_array(img)
img_array = tf.expand_dims(img_array, 0) # Create a batch

score = tf.nn.softmax(loaded_model.predict(img_array)[0])

print(
    "This image most likely belongs to {} with a {:.2f} percent confidence."
    .format(['begnin','malignant'][np.argmax(score)], 100 * np.max(score))
)
```
